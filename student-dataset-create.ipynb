## TABLES

1. Course(CourseId, Name, Credits)
2. Student(StudentId, Name)
3. Marksheet(markId, StudentId, semester, courseId, marks, fullMarks)
4. Leaves(StudentId, fromDate, toDate)

```python
import sqlite3

conn = sqlite3.connect('student.db')
cursor = conn.cursor()

# Get list of tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = [table[0] for table in cursor.fetchall()]

# Drop each table
for table in tables:
    cursor.execute(f"DROP TABLE IF EXISTS {table}")

# Commit changes and close connection 
conn.commit()
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect('student.db')
cursor = conn.cursor()

cursor.execute('''
               CREATE TABLE courses(
               id INTEGER PRIMARY KEY, 
               name TEXT, 
               department TEXT,
               semester TEXT,
               credits INTEGER
               )''')

cursor.execute('''
               CREATE TABLE students(
               id INTEGER PRIMARY KEY, 
               semester INTEGER,
               name TEXT,
               programme TEXT
               )''')

cursor.execute('''
               CREATE TABLE schedules(
               id INTEGER PRIMARY KEY,
               course_id INTEGER, 
               room TEXT NOT NULL,
               day TEXT NOT NULL,
               start_time TEXT NOT NULL,
               end_time TEXT NOT NULL,
               FOREIGN KEY (course_id) REFERENCES courses(id)
               )''')

cursor.execute('''
               CREATE TABLE registered_courses(
               id INTEGER PRIMARY KEY, 
               course_id INTEGER, 
               student_id INTEGER,
               FOREIGN KEY (course_id) REFERENCES courses(id),
               FOREIGN KEY (student_id) REFERENCES students(id)
               )''')

cursor.execute('''
               CREATE TABLE marksheets(
               id INTEGER PRIMARY KEY, 
               course_id INTEGER, 
               student_id INTEGER,
               semester INTEGER, 
               mark INTEGER, 
               full_mark INTEGER, 
               FOREIGN KEY (course_id) REFERENCES courses(id),
               FOREIGN KEY (student_id) REFERENCES students(id)
               )''')

cursor.execute('''
               CREATE TABLE leaves(
               id INTEGER PRIMARY KEY, 
               student_id INTEGER,
               from_date DATE,
               to_date DATE,
               FOREIGN KEY (student_id) REFERENCES students(id)
               )''')

conn.commit()
conn.close()
```
